```python
def draw_peak_loc(peak,outname,title_):    #peak location    figsize=(12,3)    cutoff = 0.5    bins=(12,44)    range=[[0,12],[0,44]]    figure(figsize=figsize)    h,ex,ey = np.histogram2d([v.row for k,v in peak.items()],[v.column for k,v in peak.items()],bins=bins,range=range)    h[h<cutoff]=None    imshow(h,interpolation='nearest',origin='lower')    colorbar()    title(title_)    savefig(outname,bbox_inches='tight')draw_peak_loc(sig_peak_36,'plot/sig_peak_loc_36.pdf','Signal peak location for 36cm')draw_peak_loc(dio_peak_36,'plot/dio_peak_loc_36.pdf','DIO peak location for 36cm')draw_peak_loc(sig_peak_39,'plot/sig_peak_loc_39.pdf','Signal peak location for 39cm')draw_peak_loc(dio_peak_39,'plot/dio_peak_loc_39.pdf','DIO peak location for 39cm')
```
